# Building Transform

```python
transformer = BuildingTransform()
```

```python
df_transform = pd.read_csv("./sources/input/survey_example.csv")
df_transform
```

```python
df_testing = transformer.transform(df_transform)
df_testing.head()
```

```python
features = ['address', 'person', 'habitat', 'education','religious', 'health', 'hotel', 'factory', 'otherworkspace', 'cospace']
testing = df_testing[features]
place_type = df_testing.classification.tolist()
testing.head(10)
```

```python
quality_prop = quality.placetype_proba(testing)
quality_prop
```

```python
quality.placetype_assess(testing, 0.85)
```

```python
propa = quality.proba(testing, place_type)
propa
```
